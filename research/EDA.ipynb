# Univariate Analysis

### Plotting the Distribution of Age

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.histplot(data=df, x='age', kde=True)

plt.subplot(1,2,2)
sns.histplot(data=df, x='age', kde=True, hue='gender')
plt.title('age - gender')
```

```python
sns.violinplot(data=df, x='age')
```

### Insights 
- Age is perfectly distributed evenly

### Visulaizing the Distribution of height_cm

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.histplot(data=df, x='height_cm', kde=True)
plt.title('height_cm')


plt.subplot(1,2,2)
sns.histplot(data=df, x='height_cm', kde=True, hue='gender')
plt.title('height_cm - gender')
```

### Insights
- Males heights are bigger than Females

### Visualizing the distribution of weight_kg

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.histplot(data=df, x='weight_kg', kde=True)
plt.title('weight_kg')


plt.subplot(1,2,2)
sns.histplot(data=df, x='weight_kg', kde=True, hue='gender')
plt.title('weight_kg - gender')
```

### Insights 
- Males weights are higher than Females

```python
plt.figure(figsize=(15, 6))
plt.subplot(1,2,1)
sns.countplot(data=df, x='gender', hue='gender')

plt.subplot(1,2,2)
df.gender.value_counts().plot(kind='pie')
```

### Insights 
- Males and Females counts are almost same

### Visualizing the counts of gender based on columns

```python
plt.figure(figsize=(16,7))
plt.subplot(1,3,1)
sns.countplot(data=df, x='gender', hue='activity_type')
plt.title('gender - activity_type')

plt.subplot(1,3,2)
sns.countplot(data=df, x='gender', hue='intensity')
plt.title('gender - intensity')

plt.subplot(1,3,3)
sns.countplot(data=df, x='gender', hue='smoking_status')
plt.title('gender - smoking_status')
```

### Insights
- Females are active in all kind of activites but like to do more cycling and yoga
- Males are also active in all sports but more likely to be present in weight training
- Both genders prefer activites of Medium intensity
- Both gender are more likely to not smoke
